```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
dfestudo = pd.read_csv('horas_estudo.csv', sep=';', decimal=',')
```

```python
print(dfestudo.head())
```

```python
dfestudo['horas_estudo'].mean()
```

```python
dfestudo['nota_avaliacao'].mean()
```

```python
correlation_matrix = dfestudo.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlação entre Horas de Estudo e Nota de Avaliação')
plt.show()
```

```python
#Separando as variáveis independentes
X = dfestudo[['horas_estudo']]
y = dfestudo['nota_avaliacao']
```

```python
#Criando um modelo de Regressão Linear
modelo = LinearRegression()
```

```python
#Ajustando o modelo aos dados
modelo.fit(X, y)
```

```python
#Obtendo o Intercepto
intercepto = modelo.intercept_
print(intercepto)
```

```python
#Obtendo o coeficiente angular
coeficiente_angular = modelo.coef_[0]
print(coeficiente_angular)
```

```python
#Obtendo a nota preditiva de um aluno que estudou 3 horas
horas_estudo_aluno = 3

#Calculando a nota preditiva
nota_preditiva = coeficiente_angular * horas_estudo_aluno + intercepto

print(nota_preditiva)
```

```python
#Diferença entre a nota predita e a nota real do aluno que estudou  18,06 horas

# Horas de estudo do aluno
horas_estudo_aluno = 18.06

# Calcule a nota preditiva
nota_preditiva = coeficiente_angular * horas_estudo_aluno + intercepto

print('A nota preditiva é:',nota_preditiva)

# Nota real do aluno que estudou 18,06 horas
linha_aluno = dfestudo[dfestudo['horas_estudo'] == horas_estudo_aluno]
nota_real_aluno = linha_aluno['nota_avaliacao'].values[0]
print('A nota real é:', nota_real_aluno)

#Calculando a diferença
print('A diferença entre a nota preditiva e a nota real é:', nota_real_aluno - nota_preditiva)
```

```python
#Nota preditiva de um aluno que estudou 30 horas
# Horas de estudo do aluno
horas_estudo_aluno = 30

# Calcule a nota preditiva
nota_preditiva = coeficiente_angular * horas_estudo_aluno + intercepto

print('A nota preditiva é:',nota_preditiva)
```

```python
#Socore do modelo
score_do_modelo = modelo.score(X, y)
print(score_do_modelo)
```

```python
#Obtendo o valor da MAE
# Realizando as previsões usando o modelo
previsoes = modelo.predict(X)

# Calculando o mean_absolute_error
mae = mean_absolute_error(y, previsoes)

print(mae)
```

```python
#Obtendo o valor da MSE
# Realizando as previsões usando o modelo
previsoes = modelo.predict(X)

# Calculando o mean_absolute_error
mse = mean_squared_error(y, previsoes)

print(mse)
```
